```python
import os
from keras.models import load_model
os.chdir(r"C:\Users\Toshiba\Desktop\CNN MODEL")
model = load_model('model_5.h5')
```

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
import cv2
imege = []
imgs = cv2.imread(r"C:\Users\Toshiba\Desktop\CNN MODEL\test\26.jpg")
imgs = cv2.resize(imgs,(300,300))
imgs = cv2.cvtColor(imgs,cv2.COLOR_BGR2RGB)
imege.append(imgs)
imge = np.array(imege) 
#imge = imge[...,:3]
imge = imge/255
plt.imshow(imge[0])
```

```python
pred = model.predict(imge)
pred = pred*300
pred
```

```python
A = cv2.rectangle(imge[0],(int(pred[0][0]),int(pred[0][1])),(int(pred[0][2]),int(pred[0][3])),(0,255,0),2)
plt.imshow(A)
```
