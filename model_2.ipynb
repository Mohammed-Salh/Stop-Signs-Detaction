```python
# import libiryes
import pandas as pd
import os 
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
data_csv_path = r"C:\Users\Toshiba\Desktop\CNN MODEL\bbox_v2.csv"
data_path = r"C:\Users\Toshiba\Desktop\CNN MODEL\Data"
input_dim = 300
```

```python
# loading imeges
imege = [] # make an impty list
os.chdir(data_path)
for img_path in os.listdir(data_path):
    imgs = cv2.imread(img_path) # read images from given data path
    imgs = cv2.cvtColor(imgs,cv2.COLOR_BGR2RGB) # chane color format 
    imege.append(imgs) # add imeges to the list
imge = np.array(imege) # create the array 
#imge = imge[...,:3]
imge.shape
```

```python
# creating bboxes
num_of_pics_train = 248
Y = np.zeros((num_of_pics_train,1,4))
```

```python
# loading bboxes data 
Z = pd.read_csv(data_csv_path)
i = 0
for i in range(num_of_pics_train):
    if i == 0 :
        Y[i] = Z[['xmin', 'ymin','xmax','ymax']][0:1]
    else :
        Y[i] = Z[['xmin', 'ymin','xmax','ymax']][i:i+1]
    i = i+1
```

```python
# procass & scale data 
X =  imge / 255.0 # scale input data
Y = Y.reshape(num_of_pics_train, -1) / 300 # scale target data 
Y.dtype,X.dtype 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( X, Y, test_size=0.2)
input_shape = (300,300,3)
```

```python
#beulding a model
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.optimizers import Adam
from keras import backend as K

model = Sequential()

model.add(Conv2D(32, kernel_size=(4,4),activation='relu', input_shape=input_shape))
model.add(Conv2D(32, kernel_size=(4,4),activation='relu'))
#######################################################
model.add(MaxPooling2D(pool_size=(2,2)))
#####################################################
model.add(Conv2D(64, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(64, kernel_size=(3,3),activation='relu'))
#######################################################
model.add(MaxPooling2D(pool_size=(2,2)))
#####################################################
model.add(Conv2D(128, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(128, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(128, kernel_size=(3,3),activation='relu'))
###########################################################
model.add(MaxPooling2D(pool_size=(2,2)))
##########################################################
#####################################################
model.add(Conv2D(256, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(256, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(256, kernel_size=(3,3),activation='relu'))
###########################################################
model.add(MaxPooling2D(pool_size=(2,2)))
##########################################################
#####################################################
model.add(Conv2D(512, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(512, kernel_size=(3,3),activation='relu'))
model.add(Conv2D(512, kernel_size=(3,3),activation='relu'))
###########################################################
model.add(MaxPooling2D(pool_size=(2,2)))
##########################################################
model.add(Dropout(0.1))
#########################################
model.add(Flatten())
############################################
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
############################################
model.add(Dropout(0.1))
#########################################
model.add(Dense(Y.shape[-1]))
```

```python
model.summary()
```

```python
learning_rate = 0.00005
decay_rate = learning_rate / 400
adam = Adam(lr=learning_rate,decay=decay_rate)
#sgd = SGD(lr=learning_rate,momentum=0.8,decay=decay_rate,nesterov=True)
```

```python

model.compile(
	optimizer=adam,
	loss = 'mse'
)
```

```python
model.fit( 
    x_train ,
    y_train , 
    validation_data=( x_test , y_test ),
    epochs=400 ,
    batch_size=1)
```
